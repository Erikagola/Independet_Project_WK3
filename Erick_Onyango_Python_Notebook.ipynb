# Preparation and Cleaning

```
# going through the data sets are renaming the wrongly captured names
#
dataset = dataset.rename(columns={'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'})
dataset
```

```
#renaming the inconsistent columns in dataset1
dataset1 = dataset1.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'})
dataset1
```

```
#Renaming the inconsistent columns in dataset2
dataset2 = dataset2.rename(columns={'CELLID':'CELL_ID','SIET_ID':'SITE_ID'})
dataset2
```

```
#Joining The three Data Sets 
#
telcom_data = pd.concat([dataset,dataset1,dataset2],ignore_index=True,axis=0)
telcom_data
```

```
#Checking for null and missing data in the combined dataset
telcom_data.isnull().sum()
```

```
telcom_data[telcom_data.SITE_ID.isnull()]
```

```
#Dropping rows with missing values in the telcom_data
telcom_data = telcom_data.dropna(how='any')
telcom_data
```

```
#Droppig Unamed columns in the geo_data
#
geodata.drop('Unnamed: 0',axis= 1,inplace=True)

```

```
#Removing all the null fields 
geodata = geodata.dropna(how='any')
geodata
```

```
#Removing non relevant column
telcom_data.drop('COUNTRY_B',axis=1 ,inplace=True)
```

```
#Removing non relevant column
telcom_data.drop('CELL_ON_SITE', axis=1, inplace=True)
```

```
# changing the data type on the Date column from string
#Setting it to Date Type
telcom_data['DATE_TIME']= pd.to_datetime(telcom_data['DATE_TIME'])
```
